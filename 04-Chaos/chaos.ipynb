```python
### Import Required Modules (shouldn't need to change) ###
import numpy as np                 #numpy for math
import matplotlib.pyplot as plt    #matplotlib for plotting
import cartopy.crs as ccrs         #cartopy for plotting on map
import cartopy.feature as cfeature #cartopy basic shapefiles
```

```python
## Constants ##
a = 0.25
b = 4.0
F = 8.0
G = 1.0
time_step = (4/(5*24))
```

```python
## Timestep Equation ##
def tstep_eqn(x, y, z):
    dxdt = (-(y**2) - (z**2) - (a*x) + (a*F))
    tstep_x = x + (time_step*(dxdt))
    
    dydt = ((x*y)- (b*x*z) - y + G)
    tstep_y = y + (time_step*(dydt))

    dzdt = ((b*x*y) + (x*z) - z)
    tstep_z = z + (time_step*(dzdt))
    
    return([tstep_x, tstep_y, tstep_z])
```

```python
## First Simulation ##
first_simx = np.empty([721])
first_simy = np.empty([721])
first_simz = np.empty([721])

first_simx[0] = 2.5
first_simy[0] = 1.0
first_simz[0] = 0.0

for i in range(1, 721):
    run = tstep_eqn(first_simx[i-1], first_simy[i-1], first_simz[i-1])
    first_simx[i] = run[0]   
    first_simy[i] = run[1]
    first_simz[i] = run[2]
```

```python
times = np.arange(0,721)*time_step*5
plt.figure(figsize=(10, 6))
plt.plot(times,first_simx, color='darkcyan')
plt.title("Circumpolar Westerly Current (X) // First Simulation")
plt.xlabel("Time (days)")
plt.ylabel("Strength")
plt.xlim(left=0, right=120) 
plt.grid(True)
plt.savefig("XPlot_FirstSimulation.png", dpi=250)
plt.show()
```
